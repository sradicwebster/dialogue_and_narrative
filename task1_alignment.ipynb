```python
import os
from nltk import word_tokenize
from nltk.tokenize import RegexpTokenizer as regex
from nltk.tokenize import TreebankWordTokenizer as twt
import codecs
import re
import numpy as np
import csv
import pandas as pd
import pickle
```

```python
def processAnnLines(lines):
    t=[]
    allowed=['Process','Task','Material']
    for line in lines:
        words=line.strip().split()
        if words[1] not in allowed:continue
        s=''
        for w in words[4:]:
            s+=w+' '
        s=s.strip()
        t.append((words[0],words[1],words[2],words[3],s))
    return t
```

```python
def keyphrase(ann,txt_line):
    tokens = list(twt().tokenize(txt_line))
    span = list(twt().span_tokenize(txt_line))
    
    key_phrase = ['O'] * len(tokens)
    type_kp = ['O'] * len(tokens)
    for i in range(len(tokens)):
        index=span[i]
        index=(index[0],index[1])

        for t in ann:
            if index[0]>=int(t[2]) and index[1]<=int(t[3]):
                type_kp[i] = t[1][0]
                if index[0]==int(t[2]):
                    key_phrase[i]='B'
                else:   
                    key_phrase[i]='I'
       
            
    train_data = [(tokens[i], key_phrase[i], type_kp[i]) for i in range(len(tokens))]
    
    return train_data
```

```python
files=os.listdir('.')
filenames=set()

if '.DS_Store' in files:
    files.remove('.DS_Store')
    
for name in files:
    
    filenames.add(name[:name.index('.')])

all_files = []
for name in filenames:

    f=open(name+'.ann','r')
    ann_lines=f.readlines()
    f.close()
    f=codecs.open(name+'.txt','r','utf-8')
    txt_line=f.read()
    f.close()

    ann=processAnnLines(ann_lines)
    train_data = keyphrase(ann,txt_line)

    
    all_files += train_data
    
    with open(name+'.csv', 'w') as f:
        writer = csv.writer(f)
        writer.writerows(train_data)
        #f.write("\n".join(train_data))
        #for item in train_data:
        #    f.write("%s\n" % item)
        
    
        #for t in train_data:
         #   line = ' '.join(str(x) for x in t)
          #  f.write(line + '\n')
```
